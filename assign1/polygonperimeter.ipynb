```python
import cv2
img = cv2.imread('call.png')
# resized = cv2.resize(img, (10, 10), cv2.INTER_LINEAR)
print (img)
```

```python
img = img.transpose(2,0,1).reshape(3,-1)
```

```python
img = img[0]
```

```python
img
```

```python
w, h = 28, 28;
mat = [[0 for x in range(w)] for y in range(h)] 
```

```python
c=0
for i in range(28):
    for j in range(28):
        if(img[c] <= 127):
            img[c] = 0
        else:
            img[c] = 1
        mat[i][j] = img[c]
        print(mat[i][j], end= '')
        c+=1
    print("")
```

```python
p = 0
for i in range(28):
    for j in range(28):
        if (mat[i][j] == 0):
            if(mat[i-1][j] == 1):
                p+=1
                mat[i-1][j] = 2
            if(mat[i][j-1] == 1):
                p+=1
                mat[i][j-1] = 2
            if(mat[i][j+1] == 1):
                p+=1
                mat[i][j+1] = 2
            if(mat[i+1][j] == 1):
                p+=1
                mat[i+1][j] = 2
print(p)
```
